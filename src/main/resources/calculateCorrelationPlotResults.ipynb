```python
import matplotlib.pyplot as plt
```

```python
import numpy
```

```python
import json
 
# Opening JSON file
with open('result4.json') as json_file:
    data = json.load(json_file)
```

```python
list = data.values()

metricList = ["averageCommitSizeMetricResult","averageFileSizeMetricResult","commitCountMetricResult","contributorEntropyMetricResult","linesOfCodeMetricResult","numberOfBranchesMetricResult","numberOfContributorsMetricResult","numberOfFilesMetricResult","projectDurationMetricResult","graphSize","complexity","coupling", "cohesion"]

for metric in metricList:
    metricAList = []
    metricBList = []
    for item in list:
        metricAList.append(item["graphSize"])
        metricBList.append(item[metric])
    print(metric + "\n")    
    correlation = numpy.corrcoef(metricAList, metricBList)[0][1]
    print(f'pearson correlation: {correlation} \n')    
    
    
    coef = numpy.polyfit(metricAList, metricBList, 1)
    poly1d_fn = numpy.poly1d(coef)
    plt.plot(metricAList, metricBList, 'yo', metricAList, poly1d_fn(metricAList), '--k')
    plt.plot(metricAList, metricBList, linestyle='none', marker='o')
    plt.show()
    
```
