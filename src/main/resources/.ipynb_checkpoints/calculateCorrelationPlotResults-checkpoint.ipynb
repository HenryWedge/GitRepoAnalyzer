```python
import matplotlib.pyplot as plt
```

```python
import numpy
```

```python
import json
 
# Opening JSON file
with open('result.json') as json_file:
    data = json.load(json_file)
```

```python
list = data.values()

metricList = ["averageCommitSizeMetricResult","averageFileSizeMetricResult","commitCountMetricResult","contributorEntropyMetricResult","linesOfCodeMetricResult","numberOfBranchesMetricResult","numberOfContributorsMetricResult","numberOfFilesMetricResult","projectDurationMetricResult","graphSize","complexity","coupling"]

for metric in metricList:
    metricAList = []
    metricBList = []
    for item in list:
        metricAList.append(-item['coupling'])
        metricBList.append(item[metric])
    print(metric + "\n")    
    print("pearson correlation: ${numpy.corrcoef(metricAList, metricBList)[0][1]} \n")    
    plt.plot(metricAList, metricBList, linestyle='none', marker='o')
    plt.show()
```
